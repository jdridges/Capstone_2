### removing name irregularities in dataframe

```python
cust_list['name'].to_string()
```

```python
cust_list['name'] = cust_list['name'].str.replace('[^a-zA-Z0-9-. ]', '', regex = True)
```

```python
cust_list['phone'] = cust_list['phone'].str.replace('[^0-9]', '', regex = True)
cust_list['phone']
```

```python
cust_list
```

```python
def formatted_phone(phone):
    digits = str(phone)
    return f'{digits[0:3]}-{digits[3:6]}-{digits[6:]}'

cust_list['phone'] = cust_list['phone'].apply(formatted_phone)
cust_list
```
