### Formatting phone number using NNN-NNN-NNNN

First, I used regex once again to remove any characters other than numbers and converted to string format:

```python
cust_list['phone'] = cust_list['phone'].str.replace('[^0-9]', '', regex = True)
cust_list['phone']
```

Making sure it looks good in customer list...

```python
cust_list
```

Next, I created a function to add hyphons in the correct places to fit format requirements. I then applied the function to the phone column of my cust_list DataFrame:

```python
def formatted_phone(phone):
    digits = len(str(phone))
    phone_str = str(phone)
    if digits == 10:
        return f'{phone_str[0:3]}-{phone_str[3:6]}-{phone_str[6:]}'
    elif digits == 11:
        return f'{phone_str[1:4]}-{phone_str[4:7]}-{phone_str[7:]}'
   

cust_list['phone'] = cust_list['phone'].apply(formatted_phone)
cust_list
```

### Replacing NaN values in sms-opt-out
